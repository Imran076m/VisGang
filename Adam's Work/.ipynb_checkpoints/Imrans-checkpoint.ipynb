```python
import plotly.graph_objects as go
import dash
import dash_core_components as dcc
import dash_html_components as html
import plotly.express as px
import pandas as pd
import plotly.graph_objects as go
import plotly.express as px
```

```python
df = pd.read_csv(r"C:\Users\Adaml\OneDrive - TU Eindhoven\University\Year 2\Q2\Visualizations\Git\VisGang\Adam's Work\FinalDataset.csv")
df
```

```python
df = df[df.latitude != '?']
df = df[df.longitude != '?']
df['Latitude'] = pd.to_numeric(df['latitude'])
df['Longitude'] = pd.to_numeric(df['longitude'])
dfs = df
df
```

```python
fig = px.scatter_geo(df, lat="latitude", lon="longitude", color="casualty_severity", 
                     hover_name="accident_index", size="casualty_severity", projection="natural earth", 
                     scope = "europe")
fig.show()
```

```python
fig = px.density_mapbox(df, lat='latitude', lon='longitude', z='casualty_severity', radius=3,
                        center=dict(lat=54.5, lon=-3.943646), zoom=4.2, range_color=(0, 20),
                        mapbox_style="stamen-terrain")

fig.update_layout(
        title = 'Accidents in 2009',
        geo_scope='europe', 
    )

fig.update_layout(width=600, height=400, margin={"r":0,"t":0,"l":0,"b":0})
fig.update_layout(
        geo = dict(
            projection_scale=8, 
            center=dict(lat=54.5, lon=-3.943646),
        ))

app = dash.Dash()
app.layout = html.Div([
    dcc.Graph(figure=fig)
])

app.run_server(debug=True, use_reloader=False) 
```
